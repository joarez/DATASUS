```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

DataSUS = pd.read_csv('C:/Users/joare/OneDrive/Documentos/Aulas Cientista de DADOS/Anaconda/Modulo 6/sinasc_RO_2019.csv')
DataSUS.head(10)
```

#### 2) Categorize a variável APGAR5 conforme as seguintes quebras:

- 0 a 3: 'asfixia grave'
- 4 a 5: 'asfixia moderada'
- 6 a 7: 'asfixia leve'
- 8 a 10: 'normal'

```python
DataSUS.loc[(DataSUS['APGAR5'] >= 8) & (DataSUS['APGAR5'] ==10), 'Classificação_Asfixia'] = 'Normal'
DataSUS.loc[(DataSUS['APGAR5'] >= 6) & (DataSUS['APGAR5'] <=7 ), 'Classificação_Asfixia'] = 'asfixia leve'
DataSUS.loc[(DataSUS['APGAR5'] >= 4) & (DataSUS['APGAR5'] <=5 ), 'Classificação_Asfixia'] = 'asfixia moderada'
DataSUS.loc[(DataSUS['APGAR5'] >= 0) & (DataSUS['APGAR5'] <=3 ), 'Classificação_Asfixia'] = 'asfixia severa'
DataSUS.head()
```

```python
fig, axes = plt.subplots() 
plt.subplots_adjust(wspace=0.4, hspace=0.2)
sns.countplot(x='Classificação_Asfixia', data=DataSUS)
plt.ylabel('Frequencia')
plt.show()
```

#### 3) Monte um gráfico comparando a distribuição de APGAR5 por sexo para cada grupo de tempo de gestação (GESTACAO).

```python
sns.scatterplot(x = 'SEXO'
                , y = 'GESTACAO'
                , data = DataSUS
                , alpha = .5 
                , hue = 'Classificação_Asfixia'
                , size = 'GESTACAO'
                )
plt.show()
```

####  4) Monte uma tabela cruzada com APGAR5 nas colunas e GESTACAO nas linhas

```python
TabelaCruzada = pd.crosstab(DataSUS['GESTACAO'], DataSUS['APGAR5'])
print('Junção da tabela GESTAÇÃO e, a nota do APGAR')
TabelaCruzada.head()
```

```python
print('Junção da tabela GESTAÇÃO e, a nota do APGAR\n',TabelaCruzada.head())
```

####  5) Da tabela cruzada do intem anterior, calcule os percentuais de linha. Ou seja, a distribuição da variável APGAR5 para cada classe de GESTACAO dado.

```python
TabelaPercentual = TabelaCruzada.div(TabelaCruzada.sum(axis=1),axis=0)
print('Percentual para cada nota do APGAR')
TabelaPercentual
```
